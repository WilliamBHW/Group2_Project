```python
# Initial imports
import pandas as pd
from path import Path
```

```python
file = Path('../Resources/conposcovidloc.csv')
file2 = Path('../Resources/vaccines_by_age_phu.csv')
```

```python
covid_df = pd.read_csv(file, index_col=0)
vaccine_df = pd.read_csv(file2)
covid_df
```

```python
covid_df2 = covid_df
covid_df2 = covid_df2[['Accurate_Episode_Date', 'Reporting_PHU_ID', 'Reporting_PHU_Latitude', 'Reporting_PHU_Longitude']]
covid_df2
```

```python
covid_df2['Year'] =  pd.to_datetime(covid_df2.Accurate_Episode_Date).dt.strftime('%Y')
covid_df2['Month'] = pd.to_datetime(covid_df2.Accurate_Episode_Date).dt.strftime('%m')
covid_df2['Day'] = pd.to_datetime(covid_df2.Accurate_Episode_Date).dt.strftime('%d')
print(covid_df2.shape)
covid_df2.head()
```

```python
y_df = covid_df2.groupby(['Year', 'Month', 'Reporting_PHU_ID']).count()['Day'].reset_index()
y_df = pd.DataFrame(y_df)
y_df.rename(columns={'Day': 'Number_of_Case'}, inplace=True)
y_df
```
